```python
# Directory where we saved the fine-tuned checkpoint and events files
TRAIN_DIR = '/mnt/lvmhdd/tanfulun/workspaces/Data/FashionAI/tf_format_dataset/fashionai_dataset/result/train'

# Directory where the testSet data resides
EVAL_DATA_DIR='/mnt/lvmhdd/tanfulun/workspaces/Data/FashionAI/tf_format_dataset/fashionai_dataset/tfrecord_format'

# Directory where to save the evaluation events files.
TRAIN_EVAL_DIR='/mnt/lvmhdd/tanfulun/workspaces/Data/FashionAI/tf_format_dataset/fashionai_dataset/result/train_eval'
VAL_EVAL_DIR='/mnt/lvmhdd/tanfulun/workspaces/Data/FashionAI/tf_format_dataset/fashionai_dataset/result/val_eval'
```

```python
# run evaluation on training-set
os.system('bazel build //inception:flowers_eval')
os.system('bazel-bin/inception/flowers_eval \
  --eval_dir=%s \
  --data_dir=%s \
  --subset=train \
  --num_examples=71762 \
  --checkpoint_dir=%s \
  --input_queue_memory_factor=8 \
  --run_once'%(TRAIN_EVAL_DIR,EVAL_DATA_DIR,TRAIN_DIR))
```
